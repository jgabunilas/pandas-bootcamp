## Skill Challenge - Nulls and Non-Nulls
Instructions


1.   Isolate the non-nulls in the *wine* series and assign them to the variable wine_servings
2.   Calculate the total amount of wine consumed by all countries in *wine_servings*
3. Calculate the total wine consumed by countries that consumed less than 100 servings.
* Hint: apply a boolean mask to identify such countries, then `sum()`



```python
import pandas as pd
```

Read in the CSV, keeping only the **wine_servings** and **country** columns and setting **squeeze** to `True`. Set **country** as the index column

```python
wine = pd.read_csv('https://gist.githubusercontent.com/sh7ata/e075ff35b51ebb0d2d577fbe1d19ebc9/raw/b966d02c7c26bcca60703acb1390e938a65a35cb/drinks.csv', usecols=['wine_servings', 'country'], index_col='country', squeeze=True)
```

Check the read-in

```python
wine
```

Isolate the non-null values

```python
wine_non_nulls = wine[wine.notnull()]
wine_non_nulls
```

Sum the total number of wine servings consumed by all countries that served wine.

```python
wine_non_nulls.sum()
```

To find the total wine consumed by countries with over 100 servings, use a conditional statement with the `loc` method

```python
under_100_wines = wine_non_nulls.loc[wine_non_nulls < 100]
under_100_wines
```

```python
under_100_wines.sum()
```

Another approach is to use a boolean mask. First use a conditional to obtain a boolean series of whether or not the countries that served wine served less than 100 servings

```python
under_100_wines_bool = wine_non_nulls < 100
under_100_wines_bool
```

Now apply the mask to the list of wine-serving countries, and then sum up those servings! The answer is the same as above.

```python
under_100_wines[under_100_wines_bool].sum()
```
