# Series Sorting Skill Challenge

1. Select all countries that have more than 50 wine servings, and save them (and their corresponding values) to a variable called `fifty_plus`
2. From `fifty_plus`, choose the countries with the smallest 20 wine servings values.
3. Find the mean, median, and standard deviation from the sample from Step 2.



```python
import pandas as pd
```

Import the data

```python
wine = pd.read_csv('https://gist.githubusercontent.com/sh7ata/e075ff35b51ebb0d2d577fbe1d19ebc9/raw/b966d02c7c26bcca60703acb1390e938a65a35cb/drinks.csv', usecols=['country','wine_servings'], index_col = 'country', squeeze=True)
wine
```

Select the countries with more than 50 servings

```python
fifty_plus = wine[wine > 50]
print(fifty_plus)
print(fifty_plus.count())
```

From these, select the countries with the 20 smallest wine servings

```python
twenty_smallest = fifty_plus.nsmallest(20)
twenty_smallest
```

Calculate the mean, median, and standard deviation


```python
print(twenty_smallest.mean())
print(twenty_smallest.median())
print(twenty_smallest.std())
```
