# Section 4 Skill Challenge: Extracting Data from DataFrames

Instructions:
1. Randomly select 10 food items and assign the resulting dataframe to a new variable called `nutr_mini`
2. From `nutri_mini`, extract the `total_fat` and `cholesterol` columns for all rows.
3. Extract all columns from **vitamin_b12** to the end, for the first, second, and third rows only.
4. Get the calories for the third food in *nutri_mini* using an attribute-based approach that is faster than `.loc` or `.iloc`.


Import the data

```python
import pandas as pd
```

```python
dataurl = 'https://andybek.com/pandas-nutrition'
nutrition = pd.read_csv(dataurl)
nutrition.head(10)
```

Select 10 random items and assign to *nutr_mini*

```python
nutr_mini = nutrition.sample(10)
nutr_mini
```

Set the index of the DataFrame as the *name* column

```python
nutr_mini.set_index('name', inplace=True)
nutr_mini
```

Drop the **Unnamed: 0** column using the `drop()` method. Be sure to specify the axis as `1`

```python
nutr_mini.drop(labels = 'Unnamed: 0', axis = 1, inplace=True)
nutr_mini
```

Extract the **total_fat** and **cholesterol** columns for all rows. Here we will use `.loc` and pass in two lists: the first specifies the rows (foods) and the second specifies the columns (*total_fat* and *cholesterol*)

```python
nutr_mini.loc[
              :,
              ['total_fat','cholesterol']
              ]
```

Extract all columns from **vitamin_b12** until the end for the first, second, and third rows.

Start by using get_loc() to identify the position of the **vitamin_b12** column

```python
nutr_mini.columns.get_loc('vitamin_b12')
```

Now use this to slice the columns from that position to the end while also slicing the first three rows. Remember Python indexing rules mean the rows slice will go from indices 0 to 3.

```python
nutr_mini.iloc[
               0:3,
               20:
]
```

Note that an alternative approach could have been to use the `get_loc()` method within the actual `iloc` call. The result is exactly the same.

```python
nutr_mini.iloc[
               0:3,
               nutr_mini.columns.get_loc('vitamin_b12'):
]
```

Finally, get the calories for the third food in nutr_mini using attribute-based approaches. 

We will use `.at` for this approach. We know the third food is indexed at `2`, and we want the *calories* column. 

```python
nutr_mini.at[nutr_mini.index[2], 'calories']
```

Alternatively, we could have used the `.iat` attribute. Agaiin the third food is indexed at row 2, and the calories column is at the second position, indexed at `1`. However, instead of looking for it, we can also use the `get_loc()` method within `.iat`. The result is exactly the same.

```python
nutr_mini.iat[2, nutr_mini.columns.get_loc('calories')]
```
